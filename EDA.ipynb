### Data Visualisation

````python
sns.distplot(gold_data['Adj Close'], kde = False, bins =20)

plt.show()
````

````python
sns.scatterplot('Adj Close', 'Open', data = gold_data)
plt.show()
````

````python
plt.figure(figsize=(10, 5))
sns.scatterplot('Adj Close', 'Close', data = gold_data)
plt.show()
````

````python
sns.scatterplot('Adj Close', 'High', data = gold_data)
plt.show()
````

````python
sns.scatterplot('Adj Close', 'Low', data = gold_data)
plt.show()
````

````python
sns.scatterplot('Adj Close', 'Volume', data = gold_data)
plt.show()
````

````python
sns.jointplot('Adj Close', 'Close', data = gold_data, kind = 'hex')
plt.show()
````

````python
df = gold_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
df.head()
````

````python
sns.pairplot(df)
plt.show()
````

````python
df.corr()
````

````python
sns.heatmap(df.corr(), cmap = 'coolwarm', annot = True)
plt.show()
````

````python
sns.lmplot(x = 'Adj Close', y = 'Close', data = df)
plt.show()
````

````python
sns.lmplot(x = 'Adj Close', y = 'Volume', data = df)
plt.show()
````

````python
plt.figure(figsize = (10, 8))
sns.lineplot(x = 'Date', y = 'Adj Close', data = gold_data.tail(10))

plt.show()
````
